**9) Create a text-based decision-making game. The game should present players with scenarios where choices affect the storyline and score. Multiple story paths and endings must exist. The system should track decisions, compute scores, and allow replayability. String processing and conditional branching are core requirements.**

```python
# Text-Based Decision Making Game
# Beginner-friendly Python program

def start_game():
    score = 0
    decisions = []

    print("\nðŸŒ² Welcome to THE FORGOTTEN FOREST ðŸŒ²")
    print("Your choices will shape your destiny.\n")

    # Scenario 1
    print("You wake up in a dark forest.")
    print("Do you go 'left' toward the river or 'right' toward the cave?")
    choice1 = input("> ").lower()

    if choice1 == "left":
        score += 10
        decisions.append("river")
        print("\nYou walk toward the river and find fresh water.")
    elif choice1 == "right":
        score += 5
        decisions.append("cave")
        print("\nYou enter a cave and find old markings on the walls.")
    else:
        print("\nYou hesitate too long and lose time.")
        score -= 5
        decisions.append("hesitate")

    # Scenario 2
    print("\nA stranger appears and offers help.")
    print("Do you 'trust' the stranger or 'ignore' them?")
    choice2 = input("> ").lower()

    if choice2 == "trust":
        score += 15
        decisions.append("trust")
        print("\nThe stranger gives you supplies.")
    elif choice2 == "ignore":
        score += 0
        decisions.append("ignore")
        print("\nYou move on alone, cautious but safe.")
    else:
        score -= 5
        decisions.append("confused")
        print("\nYour confusion costs you energy.")

    # Scenario 3 (branching based on earlier choice)
    if "river" in decisions:
        print("\nYou reach a broken bridge over the river.")
        print("Do you 'cross' it or 'rest' by the river?")
        choice3 = input("> ").lower()

        if choice3 == "cross":
            score += 20
            print("\nYou cross successfully and find a village.")
        else:
            score += 5
            print("\nYou rest and regain strength.")
    else:
        print("\nInside the cave, you hear a growl.")
        print("Do you 'run' or 'hide'?")
        choice3 = input("> ").lower()

        if choice3 == "hide":
            score += 20
            print("\nThe danger passes. You find hidden treasure!")
        else:
            score -= 10
            print("\nYou escape, but get injured.")

    # Ending based on score
    print("\nðŸ GAME OVER ðŸ")
    print(f"Your final score: {score}")

    if score >= 40:
        print("ðŸŒŸ Ending: Hero of the Forest")
    elif score >= 20:
        print("ðŸ™‚ Ending: Survivor")
    else:
        print("ðŸ’€ Ending: Lost Wanderer")

    # Replay option
    replay = input("\nWould you like to play again? (yes/no): ").lower()
    if replay == "yes":
        start_game()
    else:
        print("\nThanks for playing!")

# Start the game
start_game()
```

**3) Campus Store Inventory and Billing System
Develop a console-based Inventory and Billing System for a campus store. The program should allow administrators to add products, update stock levels, and define
prices. Customers should be able to purchase multiple products in a single transaction. The system must automatically generate bills showing item-wise cost, subtotal, tax, discount (based on slabs), and final payable amount. Stock levels should update dynamically, and low-stock alerts should be displayed. Negative stock values must be prevented through validation logic. Use dictionaries for inventory management and lists for cart handling.**

```python
# Campus Store Inventory and Billing System

inventory = {}
LOW_STOCK_LIMIT = 5
TAX_RATE = 0.05

def add_product():
    name = input("Enter product name: ")
    price = float(input("Enter price: "))
    stock = int(input("Enter stock quantity: "))

    inventory[name] = [price, stock]
    print("Product added successfully!\n")

def update_stock():
    name = input("Enter product name: ")
    if name in inventory:
        stock = int(input("Enter new stock quantity: "))
        if stock >= 0:
            inventory[name][1] = stock
            print("Stock updated successfully!\n")
        else:
            print("Stock cannot be negative!\n")
    else:
        print("Product not found!\n")

def show_inventory():
    print("\nAvailable Products:")
    print("Product\t\t\tPrice\t\t\tStock")
    for name in inventory:
        price, stock = inventory[name]
        print(f"{name}\t\t\t{price}\t\t\t{stock}")

        if stock <= LOW_STOCK_LIMIT:
            print("** Low Stock Alert! **")
    print()

def calculate_discount(subtotal):
    if subtotal >= 1000:
        return subtotal * 0.10
    elif subtotal >= 500:
        return subtotal * 0.05
    else:
        return 0

def customer_purchase():
    cart = []
    subtotal = 0

    while True:
        show_inventory()
        name = input("Enter product name to buy (or 'done' to finish): ")

        if name.lower() == 'done':
            break

        if name not in inventory:
            print("Product not found!\n")
            continue

        qty = int(input("Enter quantity: "))

        price, stock = inventory[name]

        if qty <= 0:
            print("Quantity must be positive!\n")
        elif qty > stock:
            print("Not enough stock available!\n")
        else:
            cart.append([name, qty])
            inventory[name][1] -= qty
            subtotal += price * qty
            print("Added to cart!\n")

    if not cart:
        print("No items purchased.\n")
        return

    discount = calculate_discount(subtotal)
    tax = subtotal * TAX_RATE
    final_amount = subtotal + tax - discount

    print("\n------- BILL -------")
    print("Item\tQty\tPrice\tTotal")
    for item in cart:
        name, qty = item
        price = inventory[name][0]
        print(f"{name}\t{qty}\t{price}\t{price * qty}")

    print("--------------------")
    print("Subtotal:", subtotal)
    print("Tax (5%):", tax)
    print("Discount:", discount)
    print("Final Amount:", final_amount)
    print("--------------------\n")

def main_menu():
    while True:
        print("1. Add Product (Admin)")
        print("2. Update Stock (Admin)")
        print("3. Show Inventory")
        print("4. Customer Purchase")
        print("5. Exit")

        choice = input("Enter your choice: ")

        if choice == '1':
            add_product()
        elif choice == '2':
            update_stock()
        elif choice == '3':
            show_inventory()
        elif choice == '4':
            customer_purchase()
        elif choice == '5':
            print("Thank you for using Campus Store System!")
            break
        else:
            print("Invalid choice!\n")

# Run the program
main_menu()
```
